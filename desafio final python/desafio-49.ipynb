```python
import matplotlib.pyplot as plt
import csv
from fpdf import FPDF

planilha = df
x = planilha.Data
y = planilha.Casos



df = pd.read_csv("Dados-covid-19-estado.csv", encoding = "ISO-8859-1", delimiter=';')



plt.plot(x, y, color = 'g', linestyle = 'dashed',
marker = 'o',label = "Dados-covid-19-estado")

plt.xticks(rotation = 25)
plt.xlabel('Data')
plt.ylabel('Casos')
plt.title('Casos de covid', fontsize = 20)
plt.grid()
plt.legend()
plt.show()

fig, ax = plt.subplots()
ax.plot(x, y, color = 'g', linestyle = 'dashed',
marker = 'o',label = "Dados-covid-19-estado")

plt.savefig('grafico_pt1.png')

pdf = FPDF('p','mm', 'a4')


# add folha
pdf.add_page()

# estilo = B, I E U
pdf.set_font('arial','',16)

#align = r = direita, l = esquerda, c = center, j = justificado
#pdf.cell(w=0,h=0,align='R', txt='texto qualquer')

texto = 'Grafico dos casos de covid'

pdf.multi_cell(w=0,h=7, align='J', txt=texto)

# imagem
pdf.image(name='grafico_pt1.png', x=0, y=20,w=200)

pdf.output('pdf1.pdf')

print('pdf criado')
```
